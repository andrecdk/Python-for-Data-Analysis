Parameter dari inclusive:

- left: min_value <= series
- right: series <= max_value
- neither: series < min_value or max_value > series

> [*n.b.*]: untuk kasus diatas dimana min. valuenya 0 dan max. valuenya 100 dan dataframe atau dataset memiliki nilai minus. Maka:

- <code>series.between(0, 100, inclusive='left')</code>, akan memasukkan semua nilai positif dan negatif sebagai nilai yang benar.
- <code>series.between(0, 100, inclusive='right')</code>, akan memasukkan nilai positif hingga 100 dan mengecualikan nilai negatif sebagai nilai yang benar.
- <code>series.between(0, 100, inclusive='neither')</code>, akan memasukkan nilai kurang dari 0 dan nilai positif lebih besar dari 100 sebagai nilai yang benar.
- <code>series.between(0, 100, inclusive='both')</code>, akan memasukkan nilai dari 0 sampai 100 sebagai nilai yang benar.

> <code>_[n.b.]</code>:_

Komponen dari regex:

- <code>.</code>: Cocok dengan karakter apa pun (kecuali baris baru).
- <code>\*</code>: Mencocokkan nol atau lebih kemunculan karakter sebelumnya.
- <code>+</code>: Cocok dengan satu atau lebih kemunculan karakter sebelumnya.
- <code>?</code>: Cocok dengan nol atau satu kemunculan karakter sebelumnya.
- <code>[...]</code>: Kelas karakterâ€”cocok dengan karakter apa pun di dalam tanda kurung.
- <code>\d</code>: Cocok dengan digit apa pun (0-9).
- <code>\w</code>: Cocok dengan karakter kata apa pun (huruf, angka, garis bawah).
- <code>\s</code>: Cocok dengan karakter spasi apa pun (spasi, tab, baris baru).
- <code>^</code>: Cocok dengan awal string.
- <code>$</code>: Cocok dengan akhir string.

#### **3.3.3. Mengelompokkan data menggunakan Pernyataan Kasus (_Case Statement_)**


##### **a. Menggunakan pandas**


```python
df6= df5.copy()

def categorize_product(row):
    product_name = row['product_name'].lower()
    price=row['price']
    
    if 'tea' in product_name:
      if 0 <= price <= 2:
        return 'cheap tea'
      elif 2 <= price <= 4:
        return 'regular tea'
      else:
        return 'expensive tea'
    
    elif 'coffee' in product_name:
      if 0 <= price <= 2:
        return 'cheap coffee'
      elif 2 <= price <= 4:
        return 'regular coffee'
      else:
        return 'expensive coffee' 
      
    else:
      return 'undetermined'

df6['product_class']= df6.apply(categorize_product, axis=1)

df6.head()
```

> Mendapatkan ringkasan setelah mengelompokkan data dengan pernyataan kasus


```python
df6_summary = (
   df6
   .groupby('product_class', observed=True)
   .agg(count_product=('product_code','count'), 
        min_price=('price','min'), 
        max_price=('price','max'))
)
df6_summary = (
    df6
    .groupby('product_class', observed=True)
    .agg(count_product=('product_code', 'count'), 
         min_price=('price', 'min'), 
         max_price=('price', 'max'))
)

df6_summary['price_intervals']= (
     df6_summary['max_price'] 
     - df6_summary['min_price']
)
df6_summary
```

##### **b. Menggunakan numpy**

> <code>np.select()</code>


```python
df7 = df6.copy()
df7['product_class2'] = np.select(
    [(df7['product_name'].str.lower().str.contains('tea')) & (df6['price'].between(0, 2, inclusive='right')),
     (df7['product_name'].str.lower().str.contains('tea')) & (df6['price'].between(2, 4, inclusive='right')),
     (df7['product_name'].str.lower().str.contains('tea')) & (df6['price']>4),
     (df7['product_name'].str.lower().str.contains('coffee')) & (df6['price'].between(0, 2, inclusive='right')),
     (df7['product_name'].str.lower().str.contains('coffee')) & (df6['price'].between(3, 5, inclusive='right')),
     (df7['product_name'].str.lower().str.contains('coffee')) & (df6['price']>5)],
    ['cheap tea', 'medium tea', 'expensive tea', 
     'cheap coffee', 'medium coffee', 'expensive coffee'],
    default='undetermined'
)
df7.head()
```

> Menggunakan Pendekatan Iteratif (_looping_)


```python
dfx=df6.copy()
def categorize_product_iterative(df):
    product_class = []  # Initialize an empty list to store the categories

    for _, row in df.iterrows():
        product_name = row['product_name'].lower()
        price = row['price']

        if 'tea' in product_name:
            if 0 <= price <= 2:
                product_class.append('Cheap Tea')
            elif 2 < price <= 4:
                product_class.append('Medium Tea')
            else:
                product_class.append('Expensive Tea')
        elif 'coffee' in product_name:
            if 0 <= price <= 2:
                product_class.append('Cheap Coffee')
            elif 3 <= price <= 5:
                product_class.append('Medium Coffee')
            else:
                product_class.append('Expensive Coffee')
        else:
            product_class.append('undetermined')

    return product_class

dfx['price_class_iterative'] = categorize_product_iterative(dfx)
dfx.head()
```

> Menggunakan pendekatan rekrusif, namun tidak disarankan pada kasus ini, hanya untuk mendemonstrasikan saja.


```python
df7=df5.copy()
def categorize_product_recursive(df, idx=0, price_class=[]):
    if idx >= len(df):
        return price_class

    product_name = df.loc[idx, 'product_name'].lower()
    price = df.loc[idx, 'price']

    if 'tea' in product_name:
        if 0 <= price <= 2:
            price_class.append('Cheap Tea')
        elif 2 < price <= 4:
            price_class.append('Medium Tea')
        else:
            price_class.append('Expensive Tea')
    elif 'coffee' in product_name:
        if 0 <= price <= 2:
            price_class.append('Cheap Coffee')
        elif 3 <= price <= 5:
            price_class.append('Medium Coffee')
        else:
            price_class.append('Expensive Coffee')
    else:
        price_class.append('Undetermined')

    return categorize_product_recursive(df, idx + 1, price_class)

df7['price_class_recursive'] = categorize_product_recursive(df7)
df7.head()
```

> Simplifikasi dari kode diatas menggunakan statmen kondisi dan <code>np.select</code> dengan metode pemograman deklaratif, bukan secara rekrusif ataupun iteratif


```python
df7 = df6.copy()
is_tea = df7['product_name'].str.lower().str.contains('tea')
is_coffee = df7['product_name'].str.lower().str.contains('coffee')

cheap_condition = df6['price'].between(0, 2, inclusive='right')
medium_condition = df6['price'].between(2, 4, inclusive='right')
expensive_condition = df6['price'] > 4

df7['price_class2'] = np.select(
    [is_tea & cheap_condition,
     is_tea & medium_condition,
     is_tea & expensive_condition,
     is_coffee & cheap_condition,
     is_coffee & medium_condition,
     is_coffee & expensive_condition],
    ['cheap tea', 'medium tea', 'expensive tea', 
     'cheap coffee', 'medium coffee', 'expensive coffee'],
    default='undetermined'
)
df7.head()
```

> Mendapatkan ringkasan dari pengelompokkan menggunakan pernyataan kasus


```python
price_class_summary = (
  df7
  .groupby('price_class2', observed=True)
  .agg(product_count=('product_code','count'),
       min_price=('price', 'min'),
       max_price=('price','max'))
)
price_class_summary['price_class_interval'] = (
  price_class_summary['max_price'] 
  - price_class_summary['min_price']
)
price_class_summary
```
