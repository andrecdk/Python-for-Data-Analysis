Langkah utama dalam proses pembersihan data meliputi 8 tahapan seperti berikut:

1. **Menghilangkan data yang tidak relevan**: mengidentifikasi dan mengeliminasi kolom atau variabel yang tidak relevan untuk analisis.
2. **Menangani data duplikat**: mengidentifikasi dan mengeliminasi duplikat _records (rows)_ dari dataset.
3. **Menangani nilai yang hilang**: menangani nilai yang hilang dengan mengimputasi nilai (mis. menggunakan mean, median atau metode lain) atau menghilangkan baris nilai yang hilang.
4. **Standarisasi Kapitalisasi Huruf**: memastikan konsistensi penggunaan huruf kapital dalam dataset.
5. **Mengonversi Tipe Data**: memastikan tipe data (mis. numerik, kategori, tanggal dan waktu) sesuai dengan analisis yang akan dilakukan.
6. **Menangani Outliers**: mendeteksi dan mengatasi nilai outliers yang menyimpang secara signifikan. Memutuskan untuk menghapus atau mentransformasikannya.
7. **Memperbaiki kesalahan**: Mendeteksi dan memperbaiki kesalahan seperti salah ketik, entri yang salah atau format yang tidak konsisten.
8. **Translasi Bahasa (Jika dapat dilakukan)**: Memastikan penggunaan bahasa yang konsisten.

Jika ingin memastikan jumlah nilai yang hilang pada setiap list dari kolom yang ada dalam tabel menggunakan fungsi <code>.isna()</code> dan juga <code>.sum()</code>


```python
# Mengecek nilai yang hilang
title = "Missing Values per Column:"
missing_values_count = new_df.isna().sum()
print(f"{title}\n{missing_values_count}")
```

atau dapat membuat kode baru untuk mengecek nilai per kolom seperti berikut:


```python
# Mendefinisikan fungsi baru untuk memunculkan jumlah nilai yang hilang pada 
# dataframe yang sudah dibersihkan dari nilai duplikat
def custom_missing_info(new_df):
    title = "Custom Missing Values Info"
    missing_values_count = new_df.isna().sum()

    # Membuat list untuk menahan modifikasi perhitungan nilai yang hilang (Create a list to hold the modified missing values count)
    missing_values_count_with_message = []

    # Ulangi jumlah nilai yang hilang dan tambahkan pesan khusus
    for column, count in missing_values_count.items():
        missing_values_count_with_message.append(f"{count} missing")

    # Buat DataFrame untuk menyimpan informasi yang diubah
    missing_df = pd.DataFrame({
        "Column": missing_values_count.index,
        "Missing Values Count": missing_values_count_with_message,
        "Data Type": new_df.dtypes.values
    })
    
    # Memformat output
    formatted_output = f"{title}:\n\n"
    formatted_output += missing_df.to_string(header=True, index=True, justify='right') 
    
    # Print output yang diformat
    print(formatted_output)
    
# contoh Penggunaan:
# Mengecek dari dataframe yang sudah dibersihkan dari nilai duplikat
custom_missing_info(new_df)
```

Mencari nilai yang hilang dengan menggunakan <code>.isna()</code>, **nilai yang hilang** pada tabel akan diberi label <code>TRUE</code>. Jika ingin menampilkan dengan cara sebaliknya, yaitu menampilkan nilai yang tidak hilang, dapat menggunakan <code>.notna()</code> nilai pada tabel yang memiliki nilai <code>FALSE</code> adalah **nilai yang hilang**


```python
# Mencari data dengan nilai yang hilang
new_df.isna() 
```

```python
# Membuat fungsi untuk mengecek data dengan nilai yang hilang pada variabel new_df
def custom_missing_reveal(new_df):
    missing_values_exist = new_df.isnull().any()
    missing_rows = new_df[new_df.isnull().any(axis=1)]

    if missing_values_exist.any():
        print(f"Behold! Missing values have revealed themselves:\n\n{missing_rows}")
    else:
        print("Huzzah! Your DataFrame is whole—no missing values in sight.")

custom_missing_reveal(new_df)
```

#### **2.5.4. Mengganti atau Menghapus nilai yang hilang**

Jika nilai yang hilang merupakan kunci dari baris atau record maka sebaiknya baris atau record data tersebut dihapus.

Untuk nilai yang hilang, terdapat beberapa pilihan seperti menghilangkan baris yang memiliki nilai yang hilang atau mengganti nilai yang hilang dengan imputasi ataupun mengganti dengan cara berikut jika datanya merupakan integer:

1. Mengganti nilai hilang dengan nilai 0
2. Mengganti nilai hilang dengan rata-rata (mean)
3. Mengganti nilai hilang dengan median
4. Mengganti nilai hilang dengan modus
5. Mengganti nilai hilang dengan nilai sebelum atau sesudah


##### **a. Menghapus nilai yang hilang**

Contoh dalam penghapusan baris dari record yang terdapat nilai yang hilang.


```python
# Membuat variabel baru untuk menampung dataframe dari baris record yang terdapat nilai yang hilang
new_df_dropna = new_df.dropna(subset=['age'], inplace=False)

new_df_dropna
```

##### **b. Mengganti nilai yang hilang dengan nilai 0**

Contoh mengganti nilai na dengan 0 menggunakan variabel dataframe baru.


```python
df_fillna = new_df.copy()
df_fillna['reward_point'] = df_fillna['reward_point'].fillna(0)
df_fillna
```

##### **c. Mengganti nilai yang hilang dengan rata-rata**

Contoh mengganti nilai na dengan rata-rata


```python
df_fillna_mean = new_df.copy()
df_fillna_mean['age'] = df_fillna_mean['age'].fillna(df_fillna_mean['age'].mean().round())
df_fillna_mean['reward_point'] = df_fillna_mean['reward_point'].fillna(df_fillna_mean['reward_point'].mean().round())
df_fillna_mean
```

##### **d. Mengganti nilai yang hilang dengan median**

Contoh mengganti nilai na dengan median


```python
df_fillna_median = new_df.copy()
df_fillna_median['age'] = df_fillna_median['age'].fillna(df_fillna_median['age'].median())
df_fillna_median['reward_point'] = df_fillna_median['reward_point'].fillna(df_fillna_median['reward_point'].median())

df_fillna_median
```

##### **e. Mengganti nilai yang hilang dengan modus**

Contoh mengganti nilai na dengan modus


```python
df_fillna_modus = new_df.copy()
df_fillna_modus['age'] = df_fillna_modus['age'].fillna(df_fillna_modus['age'].mode()[0])
df_fillna_modus['reward_point'] = df_fillna_modus['reward_point'].fillna(df_fillna_modus['reward_point'].mode()[0])
df_fillna_modus
```

##### **f. Mengganti nilai yang hilang dengan nilai sebelum dan sesudah**

Menggunakan fungsi :

- Nilai sebelum (_forward filling_):

<code>series.fillna(method'ffill')</code>

Untuk setiap nilai yang hilang (NaN), nilai tersebut akan diganti dengan nilai non-NaN sebelumnya di kolom yang sama.
Pada dasarnya, ini menyebarkan observasi valid terakhir sebelumnya ke observasi valid berikutnya.
Bayangkan ini seperti mengisi kekosongan dengan nilai terbaru yang diketahui.

- Nilai sesudah (_before filling_):

<code>series.fillna(method'bfill')</code>

Untuk setiap nilai yang hilang (NaN), ia menggantinya dengan nilai non-NaN berikutnya di kolom yang sama.
Ia menggunakan observasi valid berikutnya untuk mengisi kesenjangan tersebut.
Anggap saja sebagai mengisi kekosongan dengan nilai yang diketahui di masa depan.


```python
# Mengganti nilai na dengan nilai sebelumnya
df_fillna_ffill = new_df.copy()
df_fillna_ffill['age'] = df_fillna_ffill['age'].fillna(df_fillna_ffill['age'].mode()[0])
df_fillna_ffill['reward_point'] = df_fillna_ffill['reward_point'].fillna(df_fillna_ffill['reward_point'].mode()[0])
df_fillna_ffill
```

```python
# Mengganti nilai na dengan nilai setelahnya
df_fillna_bfill = new_df.copy()
df_fillna_bfill['age'] = df_fillna_bfill['age'].fillna(df_fillna_bfill['age'].mode()[0])
df_fillna_bfill['reward_point'] = df_fillna_bfill['reward_point'].fillna(df_fillna_bfill['reward_point'].mode()[0])
df_fillna_bfill
```

##### c. Contoh dalam menstandarisasi data yang tidak konsisten

Standarisasi hanya berlaku pada tipe data teks atau string untuk huruf besar dan huruf kecil


Contoh standarisasi untuk huruf besar

> <code>.str.upper()</code>


```python
df_uppercase = pd.DataFrame({
  'product_name':['ICE TEA','Ice Tea','ice tea']
})
print(df_uppercase)
df_uppercase['product_name'] = df_uppercase['product_name'].str.upper()
print('\n')
print(df_uppercase)
```

Contoh standarisasi huruf kecil

> <code>.str.lower()</code>


```python
df_lowercase = pd.DataFrame({'product_name':['ICE TEA','Ice Tea','ice tea']})
print(df_lowercase)
df_lowercase['product_name'] = df_lowercase['product_name'].str.lower()
print('\n')
print(df_lowercase)
```

Contoh dalam menangani extra whitespace atau karakter ekstra spasi pada data.

> <code>.str.strip()</code>


```python
df_extrawhitespace = pd.DataFrame({
  'product_name':['   ICE TEA   ','ICE COFFEE   ','    RAINBOW CAKE']
})
print(df_extrawhitespace)
df_extrawhitespace['product_name'] = df_extrawhitespace['product_name'].str.strip()
print('\n')
print(df_extrawhitespace)
```

Contoh dalam menangani teks data yang tidak perlu

> <code>series.str.replace(oldvalue, newvalue, count=0, regex=True)</code>


```python
df_extrateks= pd.DataFrame({
  'product_name':['ICE TEA (SOLD PER 1 GLASS)','RAINBOW CAKE (SOLD PER 1 SLICE)']
})
print(df_extrateks)
df_extrateks['product_name'] = df_extrateks['product_name'].str.replace(r' \(SOLD.*', '', regex=True)
print('\n')
print(df_extrateks)
```

Contoh dalam menangani teks yang tidak konsisten


```python
df_nonconsistent = pd.DataFrame({
  'address':['jln. harapan','jl. baru']
})
print(df_nonconsistent)
df_nonconsistent['address'] = df_nonconsistent['address'].str.replace(r'jl.* ', 'jalan ', regex=True)
print('\n')
print(df_nonconsistent)
```
